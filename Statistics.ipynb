#Introduction to Statistics

#Descriptive Statistics

#1.Measure of Central Tendancy

#Mean, Median, Mode, Range

```python
import pandas as pd

# Student marks datasets
marks1 = [91, 98, 97, 96, 95, 94, 95, 92]
marks2 = [60, 78, 65, 80, 95, 70, 88, 95]

# Create DataFrames
df1 = pd.DataFrame(marks1, columns=['Marks'])
df2 = pd.DataFrame(marks2, columns=['Marks'])

# Calculate statistics for marks1
mean1 = df1['Marks'].mean()
median1 = df1['Marks'].median()
mode1 = df1['Marks'].mode().tolist()

# Calculate statistics for marks2
mean2 = df2['Marks'].mean()
median2 = df2['Marks'].median()
mode2 = df2['Marks'].mode().tolist()

print(f"Marks1 - Mean: {mean1}, Median: {median1}, Mode: {mode1}")
print(f"Marks2 - Mean: {mean2}, Median: {median2}, Mode: {mode2}")

# Calculate range for marks1
range1 = df1['Marks'].max() - df1['Marks'].min()

# Calculate range for marks2
range2 = df2['Marks'].max() - df2['Marks'].min()

print(f"Marks1 - Range: {range1}")
print(f"Marks2 - Range: {range2}")
```

#Interpretation:

Mean:

Marks1: The mean is 94.5, which is the average score of the students. This indicates that on average, students scored quite high in this dataset.
Marks2: The mean is 79.875, which is lower than Marks1. This suggests that the average score in this dataset is lower.

Median:

Marks1: The median is 95.0, which is the middle value when the data is ordered. Since the median is higher than the mean, it indicates that the distribution might be slightly skewed to the right.
Marks2: The median is 80.0, which is close to the mean. This suggests a relatively symmetric distribution in this dataset.

Mode:

Marks1: The mode is 95, meaning this is the most frequently occurring score in this dataset. The presence of a mode indicates that 95 is a common score among the students.
Marks2: The mode is also 95, showing that this score is the most frequent in this dataset as well. This could indicate a clustering of scores around this value.



#2.Measure of Dispersion

#Variance and Standard Deviation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create student marks datasets
marks1 = [91, 98, 97, 96, 95, 94, 95, 92]
marks2 = [60, 78, 65, 80, 95, 70, 88, 95]

# Create DataFrames
df1 = pd.DataFrame({'Marks': marks1})
df2 = pd.DataFrame({'Marks': marks2})

# Calculate statistics for df1
mean1 = df1['Marks'].mean()
variance1 = df1['Marks'].var()
std_dev1 = df1['Marks'].std()

# Calculate statistics for df2
mean2 = df2['Marks'].mean()
variance2 = df2['Marks'].var()
std_dev2 = df2['Marks'].std()

# Print statistics
print("df1:")
print("  Mean:", mean1)
print("  Variance:", variance1)
print("  Standard Deviation:", std_dev1)

print("\ndf2:")
print("  Mean:", mean2)
print("  Variance:", variance2)
print("  Standard Deviation:", std_dev2)


```

#Data Visualization

```python
# Create subplots
fig, ax = plt.subplots(figsize=(10, 4))

# Plot distributions
sns.distplot(df1['Marks'], label='Class 1', ax=ax)
sns.distplot(df2['Marks'], label='Class 2', ax=ax)

# Add mean lines
ax.axvline(mean1, color='red', linestyle='--', label='Mean (Class 1)')
ax.axvline(mean2, color='blue', linestyle='--', label='Mean (Class 2)')

# Customize plot
ax.set_title('Distribution of Student Marks in Two Classes')
ax.legend()
plt.show()
```

#Mean vs Std

```python

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot distributions
sns.distplot(df1['Marks'], label='Class 1', ax=ax1)
sns.distplot(df2['Marks'], label='Class 2', ax=ax2)

# Add mean lines
ax1.axvline(mean1, color='red', linestyle='--', label='Mean')
ax2.axvline(mean2, color='red', linestyle='--', label='Mean')

# Add standard deviation lines
ax1.axvline(mean1 + std_dev1, color='green', linestyle='--', label='Std Dev')
ax1.axvline(mean1 - std_dev1, color='green', linestyle='--')
ax2.axvline(mean2 + std_dev2, color='green', linestyle='--', label='Std Dev')
ax2.axvline(mean2 - std_dev2, color='green', linestyle='--')

# Customize plots
ax1.set_title('Class 1: Distribution with Mean and Std Dev')
ax1.legend()
ax2.set_title('Class 2: Distribution with Mean and Std Dev')
ax2.legend()

plt.show()
```

#3.Handling Outlier in the data - 5 number summary

#Percentile , Quartile, Inter Quartile Range

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Given data
data = [1, 2, 2, 2, 3, 3, 4, 5, 5, 5, 6, 6, 6, 6, 7, 8, 8, 9, 29]

# Create a DataFrame
df = pd.DataFrame({'Values': data})

# Calculate the five number summary
min_value = df['Values'].min()
q1 = df['Values'].quantile(0.25)
median = df['Values'].median()
q3 = df['Values'].quantile(0.75)
max_value = df['Values'].max()

five_number_summary = {
    "Minimum": min_value,
    "First Quartile (Q1)": q1,
    "Median (Q2)": median,
    "Third Quartile (Q3)": q3,
    "Maximum": max_value
}

print("Five Number Summary:")
print(five_number_summary)

# Plotting the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Values')

# Adding titles and labels
plt.title('Box Plot of Given Data')
plt.xlabel('Values')

# Show plot
plt.show()
```

#4.Measure of Symmetry

#Calculating Skewness and Kurtosis


```python
import pandas as pd
from scipy.stats import skew, kurtosis

# Example data: Student Test Scores
scores = [55, 65, 70, 75, 80, 85, 90, 92, 95, 98]

# Create a DataFrame
df = pd.DataFrame({'Scores': scores})

# Calculate skewness and kurtosis
skewness = skew(df['Scores'])
kurtosis_value = kurtosis(df['Scores'])

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis_value}")
```

#Negative skew (left-skewed)Example


```python
import pandas as pd
from scipy.stats import skew, kurtosis
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Student Marks
marks = [45, 50, 55, 60, 65, 68, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100, 105]

# Create a DataFrame
df = pd.DataFrame({'Marks': marks})

# Calculate skewness and kurtosis
skewness = skew(df['Marks'])
kurtosis_value = kurtosis(df['Marks'])
# Calculate mean and median
mean_value = df['Marks'].mean()
median_value = df['Marks'].median()
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis_value}")
print("Mean:", mean_value)
print("Median:", median_value)
# Plotting histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Marks'], bins=10, kde=True)
plt.title('Distribution of Student Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show()

# Plotting box plot
plt.figure(figsize=(10,4))
sns.boxplot(data=df, x='Marks')
plt.title('Box Plot of Student Marks')
plt.xlabel('Marks')
plt.show()
```

#3 types of Kurtosis

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Create datasets with different kurtosis levels
data_lepto = [40, 45, 50, 55, 60, 65, 68, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100, 105, 120, 130]  # Leptokurtic
data_meso = [45, 50, 55, 60, 62, 65, 68, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100]  # Mesokurtic
data_platy = [60, 62, 64, 65, 66, 67, 68, 69, 70, 71, 72, 73, 74, 75, 76, 77, 78, 79, 80, 82]  # Platykurtic

# Define colors for each plot
colors = ['b', 'g', 'r']

# Calculate kurtosis for each dataset
kurtosis_lepto = kurtosis(data_lepto)
kurtosis_meso = kurtosis(data_meso)
kurtosis_platy = kurtosis(data_platy)

print("Kurtosis of leptokurtic data:", kurtosis_lepto)
print("Kurtosis of mesokurtic data:", kurtosis_meso)
print("Kurtosis of platykurtic data:", kurtosis_platy)

# Plot KDE for each dataset using seaborn
sns.kdeplot(data_lepto, label="Leptokurtic", color=colors[0])
sns.kdeplot(data_meso, label="Mesokurtic", color=colors[1])
sns.kdeplot(data_platy, label="Platykurtic", color=colors[2])

# Customize plot
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Distribution of Data with Different Kurtosis Levels")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Create datasets with different kurtosis levels
# Leptokurtic: Heavy tails, peaked distribution
data_lepto = np.concatenate([np.random.normal(0, 1, 900), np.random.normal(5, 0.5, 50), np.random.normal(-5, 0.5, 50)])

# Mesokurtic: Similar to normal distribution
data_meso = np.random.normal(0, 1, 1000)

# Platykurtic: Light tails, flat peak
data_platy = np.random.uniform(-3, 3, 1000)

# Define colors for each plot
colors = ['b', 'g', 'r']

# Calculate kurtosis for each dataset
kurtosis_lepto = kurtosis(data_lepto)
kurtosis_meso = kurtosis(data_meso)
kurtosis_platy = kurtosis(data_platy)

print("Kurtosis of leptokurtic data:", kurtosis_lepto)
print("Kurtosis of mesokurtic data:", kurtosis_meso)
print("Kurtosis of platykurtic data:", kurtosis_platy)

# Plot KDE for each dataset using seaborn
sns.kdeplot(data_lepto, label="Leptokurtic", color=colors[0])
sns.kdeplot(data_meso, label="Mesokurtic", color=colors[1])
sns.kdeplot(data_platy, label="Platykurtic", color=colors[2])

# Customize plot
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Distribution of Data with Different Kurtosis Levels")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```
